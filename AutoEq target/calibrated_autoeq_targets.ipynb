```python
from autoeq.frequency_response import FrequencyResponse
```

```python
autoeq = FrequencyResponse.read_from_csv('AutoEq in-ear.csv')
harman = FrequencyResponse.read_from_csv('Harman in-ear 2019 without bass.csv')
delta = autoeq.raw - harman.raw
```

```python
prefixes = ['Innerfidelity', 'Rtings', 'Headphone.com Legacy']
for prefix in prefixes:
    fr = FrequencyResponse.read_from_csv(f'{prefix} Harman in-ear 2019 without bass.csv')
    fr.raw += delta
    fr.name = f'{prefix} AutoEq in-ear'
    fr.write_to_csv(f'{prefix} AutoEq in-ear.csv')
    g = fr.plot_graph(file_path=f'{prefix} AutoEq in-ear.png', show=False, raw_plot_kwargs={'color': 'C0'})
```
